### Funciones

```python
def nota_final(prueba1,prueba2,prueba3,prueba4):
    promedio_pruebas = round((prueba1+prueba2+prueba3+prueba4)/4,1)
    print(promedio_pruebas)
    if promedio_pruebas > 4:
        return print('Aprobado')
    else:
        return print('Desaprobado')
    

p1 = int(input('Nota 1:'))
p2 = int(input('Nota 2:'))
p3 = int(input('Nota 3:'))
p4 = int(input('Nota 4:'))

nota_final(p1,p2,p3,p4)
```

```python
def promedio_lista(lista):

    suma = 0
    for i in lista:
        suma += i
    
    promedio = round(suma/len(lista),1)

    return (promedio)


print(promedio_lista([4,5,6,7,3,5,3,6,3,5]))
```

### ARGUMENTOS Y VARIABLES

```python
def potencia(base,exponente):

    return base ** exponente


print(potencia(3,6))
```
